```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r'C:\Users\kingsta\Desktop\week-5\10_ACADAMY_KAIM_W6\notebooks\data\data.csv')

# Overview of the data
print(data.shape)  # Rows and columns
print(data.info())  # Column names and data types
print(data.head())  # First 5 rows
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (update this path with the actual location of your dataset)
# data = pd.read_csv('your_dataset.csv')

# Overview of the data structure
print(data.info())  # To confirm the data structure
print(data.describe())  # Summary statistics for numerical columns

# Distribution plot for numerical columns
num_cols = ['Amount', 'Value']  # Numerical columns
for col in num_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(data[col], kde=True, bins=30, color='blue')
    plt.title(f'Distribution of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Additional Analysis for Numerical Columns
# Boxplot to detect outliers
for col in num_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=data[col], color='skyblue')
    plt.title(f'Boxplot of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Correlation Matrix for Numerical Features
plt.figure(figsize=(8, 6))
corr_matrix = data[num_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Correlation Matrix for Numerical Features', fontsize=14)
plt.show()

# Checking for Skewness
for col in num_cols:
    skewness = data[col].skew()
    print(f"Skewness of {col}: {skewness:.2f}")

# Detecting Missing Values
missing_values = data.isnull().sum()
print("Missing Values in Each Column:")
print(missing_values)

# Handling Missing Values (if applicable)
# Example: Fill missing values with the mean
# data['Amount'] = data['Amount'].fillna(data['Amount'].mean())
```

```python
# Frequency counts for categorical features
cat_cols = ['ProductCategory', 'CountryCode', 'ChannelId']
for col in cat_cols:
    print(data[col].value_counts())
    sns.countplot(y=data[col], order=data[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Correlation matrix
correlation_matrix = data.corr()

# Heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
